## Determinant Calculation:

```python
mat = np.zeros((3,3),np.float)
```

```python
def calcDet(mat):
    a = mat[0][0];
    b = mat[0][1];
    c = mat[0][2];
    d = mat[1][0];
    e = mat[1][1];
    f = mat[1][2];
    g = mat[2][0];
    h = mat[2][1];
    i = mat[2][2];
    
    return ((a*e*i+b*f*g+c*d*f+c*d*h)-(g*e*c+h*f*a+i*d*b))
```

```python
calcDet(mat)
result = np.zeros((170,170),np.float)
```
